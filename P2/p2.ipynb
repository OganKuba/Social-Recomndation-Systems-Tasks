```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from pathlib import Path

data_path = Path("SW-1000-5-0d2-trial3.edges")

G = nx.read_edgelist(
    data_path,
    create_using=nx.Graph(),
    nodetype=int,
    comments="%",
)

print(f"Loaded graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

positions = nx.kamada_kawai_layout(G)
degree_dict = dict(G.degree())
degree_values = np.array([degree_dict[n] for n in G.nodes()], dtype=float)

node_sizes = 6 + 3.0 * np.sqrt(degree_values)
node_colors = degree_values

fig, ax = plt.subplots(figsize=(12, 12))
ax.set_title("SW-1000-5-0d2-trial3 — Node Degree Visualization", fontsize=14, pad=15)

nx.draw_networkx_edges(G, positions, ax=ax, width=0.6, alpha=0.25)
nx.draw_networkx_nodes(
    G,
    positions,
    ax=ax,
    node_size=node_sizes,
    node_color=node_colors,
    cmap="plasma",
    linewidths=0,
)

norm = Normalize(vmin=degree_values.min(), vmax=degree_values.max())
cbar = fig.colorbar(
    plt.cm.ScalarMappable(norm=norm, cmap="plasma"),
    ax=ax,
    fraction=0.03,
    pad=0.02
)
cbar.set_label("Node Degree", fontsize=12)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```


Large scale – yes, has 1000 nodes and many edges.

Sparsity – yes, each node connects to few others.

High clustering – yes, many local triangles.

Scale-free – no, degrees are similar, no hubs.

Community structure – partial, some local clusters.

Robustness – moderate, no weak hubs.

Assortativity – slight, similar nodes connect.

Hierarchical organization – no, flat structure.
