Draw square lattice with your own implementation (not build-in
in some library).

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 5
G = nx.Graph()
pos = {(i, j): (i, j) for i in range(1, n + 1) for j in range(1, n + 1)}

G.add_nodes_from(pos.keys())

for i in range(1, n + 1):
    for j in range(1, n + 1):
        if i < n:
            G.add_edge((i, j), (i + 1, j))
        if j < n:
            G.add_edge((i, j), (i, j + 1))

nx.draw(G, pos, node_size=100, with_labels=False)
plt.show()
```

Draw chain with your own implementation (not build-in in some
library)

```python
G = nx.Graph()
G.add_edges_from((i, i + 1) for i in range(1, 12))

nx.draw(G, with_labels=True)
plt.show()
```

Draw complete graph with your own implementation (not
build-in in some library)

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 6
G = nx.Graph()
G.add_edges_from((i, j) for i in range(1, n) for j in range(i + 1, n + 1))

nx.draw(G, with_labels=True)
plt.show()
```

Draw weighted graph (any, but N > 5)

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

n = 7
G = nx.Graph()

G.add_edges_from(
    (i, j, {'weight': i + j})
    for i in range(1, n)
    for j in range(i + 1, n + 1)
    if random.random() > 0.5
)

pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))
plt.show()
```

Draw directed graph (any, but N > 5)

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

n = 7
G = nx.DiGraph()

G.add_edges_from(
    (i, j, {'weight': i + j})
    for i in range(1, n + 1)
    for j in range(i + 1, n + 1)
    if random.random() > 0.5
)

nx.draw(G, with_labels=True, arrows=True)
plt.show()
```

Compute vertex degrees for one of the above graphs

```python
print(G.degree())
```
